```python
float_patern = r"\s*([-+]?(?:\d+(?:\.\d*)?|\.\d+)(?:[eE][-+]?\d+)?)"
processor = {}
for coord in ["F", "X", "Y", "I", "J", "SPP="]:
    coord_pattern = coord + float_patern
    processor[coord_pattern] = lambda x, coord=coord: (coord + x[1], "", "ok" )

    
for pattern in processor:    
    print(processor[pattern]("dfsdfs"))
```
